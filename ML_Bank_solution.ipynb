```python
from sklearn.preprocessing import StandardScaler
num_cols = ['age', 'duration', 'campaign', 'pdays','previous','emp.var.rate','cons.price.idx','cons.conf.idx','euribor3m','nr.employed']
copy_bank_data = bank_data.copy()
scalar = StandardScaler()
copy_bank_data[num_cols] = scalar.fit_transform(copy_bank_data[num_cols])
cols_to_drop = ['job', 'marital', 'education', 'default', 'housing', 'loan', 'contact', 'month', 'day_of_week','poutcome','y']
copy_bank_data = copy_bank_data.drop(cols_to_drop, axis=1)
copy_bank_data.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)
target_bank = bank_data.copy()
columns_to_drop = ['age', 'job', 'marital', 'education', 'default', 'housing', 'loan',
                   'contact', 'month', 'day_of_week', 'duration', 'campaign', 'pdays',
                   'previous', 'poutcome', 'emp.var.rate', 'cons.price.idx',
                   'cons.conf.idx', 'euribor3m', 'nr.employed']
target_bank = target_bank.drop(columns_to_drop, axis=1)
target_bank['y'] = target_bank['y'].apply(lambda x: 1 if x == 'yes' else 0)
target_bank.head()
```

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder(sparse=False)
category_cols = ['job', 'marital', 'education', 'default', 'housing', 'loan',
                   'contact', 'month', 'poutcome']
bank_category = bank_data.copy()
df_encoded = pd.DataFrame(encoder.fit_transform(bank_category[category_cols]))
df_encoded.columns = encoder.get_feature_names_out(category_cols)
bank_category=bank_category.drop(category_cols,axis=1)
final_df=pd.concat([copy_bank_data,target_bank,df_encoded],axis=1)

final_df.shape
final_df.head()
```

**Train test Split**

```python
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, cohen_kappa_score
from sklearn import tree

# Your data preparation steps
feature = final_df.drop('y', axis=1)
target = final_df['y']
x_train, x_test, y_train, y_test = train_test_split(feature, target, shuffle=True, test_size=0.2, random_state=1)

```

Decission Tree Classifier


```python
DTC = tree.DecisionTreeClassifier(random_state=0)
DTC.fit(x_train, y_train)

# Evaluation function
def evaluate_model(model, X, y):
    y_pred = model.predict(X)
    accuracy = accuracy_score(y, y_pred)
    precision = precision_score(y, y_pred)
    recall = recall_score(y, y_pred)
    f1 = f1_score(y, y_pred)
    kappa = cohen_kappa_score(y, y_pred)

    return {'acc': accuracy, 'precision': precision, 'recall': recall, 'f1': f1, 'kappa': kappa}

# Evaluate the model using the defined evaluate_model function
DTC_eval = evaluate_model(DTC, x_test, y_test)

# Display the evaluation metrics
print('Accuracy:', DTC_eval['acc'])
print('Precision:', DTC_eval['precision'])
print('Recall:', DTC_eval['recall'])
print('F1 Score:', DTC_eval['f1'])
print('Cohen\'s Kappa:', DTC_eval['kappa'])
```

```python
from sklearn.linear_model import SGDClassifier
sgd_clf = SGDClassifier(random_state=0)
sgd_clf.fit(x_train, y_train)

def evaluate_model(model, X, y):
    y_pred = model.predict(X)
    accuracy = accuracy_score(y, y_pred)
    precision = precision_score(y, y_pred)
    recall = recall_score(y, y_pred)
    f1 = f1_score(y, y_pred)
    kappa = cohen_kappa_score(y, y_pred)

    return {'acc': accuracy, 'precision': precision, 'recall': recall, 'f1': f1, 'kappa': kappa}

sgd_eval = evaluate_model(sgd_clf, x_test, y_test)

print('SGD Classifier Metrics:')
print('Accuracy:', sgd_eval['acc'])
print('Precision:', sgd_eval['precision'])
print('Recall:', sgd_eval['recall'])
print('F1 Score:', sgd_eval['f1'])
print('Cohen\'s Kappa:', sgd_eval['kappa'])
```
